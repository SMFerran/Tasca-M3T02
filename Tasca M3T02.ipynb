```python
import pandas as pd
df= pd.read_csv ("C:\\Users\\client277192\\Documents\\01-IT ACADEMY. DATA SCIENCE\\NUMPY+PANDAS\\TASCA M3\\DelayedFlights.csv")
df
```

```python
datos_columnas= df[["Year", "Month", "UniqueCarrier","FlightNum","ActualElapsedTime", "AirTime","ArrDelay","DepDelay","Dest","Origin", "Distance","Cancelled","Diverted","CarrierDelay", "WeatherDelay","NASDelay",	"SecurityDelay","LateAircraftDelay"]]
datos =datos_columnas.sample(200000)
datos
```

```python
 #EXERCICI 2

 #1-Resumeix estadísticament el dataset i les columnes d’interès. 
 # Fes una anàlisi estadístic del que consideris rellevant
 
# Analisis- es interesante saber el maximo de retraso en la salida y llegada de los vuelos

datos["ArrDelay"] =datos["ArrDelay"].apply(lambda x: 0 if x<0 else x) #apply para aplicar una funcion personalizada a cada elemento de la columna. Lambda para una funcion simple
valor_maximo= datos[["ArrDelay", "DepDelay"]].max()
valor_maximo
```

```python
# Analisis-se muestra la media de los retrasos tanto en la llegada como partida de los vuelos, así como la media de tp de vuelo 
# y la distancia

media= round(datos[["ArrDelay", "DepDelay","AirTime", "Distance"]].mean()) #round para un rdo. sin decimales
media
```

```python
#Analisis-Se muestra el origen, destino y la compañia que han tenido mayores retrasos en las llegadas y las salidas

valor_maximo= datos[["ArrDelay", "DepDelay"]].max()
datos_ArrDelay =datos.loc[datos["ArrDelay"] == valor_maximo["ArrDelay"], ["Origin", "Dest", "UniqueCarrier", "ArrDelay"]].values[0]
print("el origen del vuelo con mas retraso en la llegada es: Origen, Destino, Compañia, Retraso") 
datos_ArrDelay
```

```python
datos_DepDelay =datos.loc[datos["DepDelay"] == valor_maximo["DepDelay"], ["Origin", "Dest", "UniqueCarrier", "DepDelay"]].values[0]
print(f"el origen del vuelo con mas retraso en la salida es: Origen, Destino, Compañia, Retraso") 
datos_DepDelay
```

```python
#Analisis-se calcula el % de vuelos cancelados y desviados
vuelos_cancelados = datos["Cancelled"]. sum()
vuelos_desviados = datos["Diverted"]. sum()
total_vuelos = datos ["FlightNum"]. count()
porcentaje_cancelados = (vuelos_cancelados/total_vuelos)*100
porcentaje_desviados = (vuelos_desviados/total_vuelos)*100
print(f"el porcentaje de vuelos cancelados es {porcentaje_cancelados:.2f}%")
print(f"el porcentaje de vuelos desviados es {porcentaje_desviados:.2f}%")
```

```python
#Analisis-se calcula el % de vuelos con retraso en la llegada teniendo en cuenta que un retraso de 15 minutos se considera como que llega a la hora
# y el % del retraso en la salida

vuelos_con_retraso_llegada =datos.loc[datos["ArrDelay"]>15, "ArrDelay"].sum() 
vuelos_con_retraso_salida = datos["DepDelay"].sum()
total_vuelos = datos ["FlightNum"].count()
porcentaje_retraso_llegada = (vuelos_con_retraso_llegada/total_vuelos)*100
porcentaje_retraso_salida = (vuelos_con_retraso_salida/total_vuelos)*100
print(f"el porcentaje de vuelos con retraso en la llegada es de {porcentaje_retraso_llegada:.2f}%")
print(f"el porcentaje de vuelos con retraso en la salida es de {porcentaje_retraso_salida:.2f}%")
```

```python
# Analisis-Calculamos que motivo produce mas retrasos, de los que estan informados
causa_retrasos= round(datos[["CarrierDelay","WeatherDelay", "NASDelay","SecurityDelay","SecurityDelay"]].sum())
causa_retrasos =causa_retrasos.sort_values(ascending=False)
causa_retrasos
```

```python
#Troba quantes dades faltants hi ha per columna. is nul

nulos = datos[["Year","Month","UniqueCarrier","AirTime","ArrDelay","DepDelay","Origin","Dest","Distance","Cancelled","Diverted","CarrierDelay","WeatherDelay","NASDelay","SecurityDelay","LateAircraftDelay"]].isnull().sum()
nulos
```

```python
#Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).
datos["Velocidad_media_vuelo"]= datos["AirTime"]/datos["Distance"]
datos["retraso"] = datos["ArrDelay"] > 0
datos.head(5)
```

```python
#Fes una taula de les aerolínies amb més endarreriments acumulats.ordeno

datos["retraso_Acumulado"]= datos["ArrDelay"] + datos["DepDelay"].cumsum() #creo culumna retraso acumulado
agrupación = datos.groupby(["UniqueCarrier"])["retraso_Acumulado"].sum()
columnas_mostrar =agrupación.reset_index()[["UniqueCarrier","retraso_Acumulado" ]]#.sort_values(by="retraso_Acumulado", ascending=False)
rdo=(columnas_mostrar.sort_values(by="retraso_Acumulado", ascending=False))
rdo
```

```python
#Quins són els vols més llargs? I els més endarrerits? 

vuelos_largos= datos.sort_values(["ActualElapsedTime"], ascending=False)
rdo1= vuelos_largos[["ActualElapsedTime", "Origin", "Dest"]]
rdo1.head(5)

```

```python

#vols mes endarrerits
datos["mas_retrasos"]= datos["ArrDelay"] + datos["DepDelay"].sum() 
vols_mes_endarrerits = datos.sort_values(["mas_retrasos"], ascending=False)
rdo2= vols_mes_endarrerits[["mas_retrasos","Origin", "Dest","FlightNum" ]]
rdo2.head(5)
```
